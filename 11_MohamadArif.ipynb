```python
def vowels(Text):
  Text = Text.lower()
  vowels = "aiueo"
  total = 0;
  charVowels = [];
  for index, char in enumerate(Text):
    if char in vowels:
      total += 1
      charVowels.append(char)

  output = f"{total} ({','.join(charVowels)})"

  return output

print(vowels("Hello World"))
print(vowels("PYTHON"))
```

```python
def count(prices):
  output = []
  total = 0
  avg = 0
  max = prices[0]
  for index, price in enumerate(prices):
    total += price
    if price > max:
      max = price

  avg = total / len(prices)
  output.append(total)
  output.append(avg)
  output.append(max)

  return tuple(output)

print(count([10, 20, 30]))
print(count([5]))
```

```python
def remDup(items):
  arrayNew = []
  for index, item in enumerate(items):
    if item not in arrayNew:
      arrayNew.append(item)

  return arrayNew

print(remDup([1, 2, 1, 3, 2]))
print(remDup(["a", "b", "a", "c", "b"]))
```

```python
from os import name
class StudentGradeBook:
  def __init__(self, name):
    self.name = name
    self.score = []

  def add_score(self, score):
    if score > 100:
      return "Invalid Score"
    else:
      self.score.append(score)
      return f"Added {score}"

  def average(self):
    if not self.score:
      return 0
    else:
      return sum(self.score) / len(self.score)

  def status(self):
    average = self.average()
    if average >= 75:
      status = "Pass"
    elif average >= 60:
      status = "Remedial"
    else:
      status = "Fail"
    return status
```

```python
s = StudentGradeBook("Arif")
print(s.average())
print(s.status())
print(s.add_score(80))
print(s.add_score(70))
print(s.add_score(120))
print(s.average())
print(s.status())
```
